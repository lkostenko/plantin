##СРА

```python
marketing_spend = {
    '2024-03': 180000,
    '2024-04': 380000,
    '2024-05': 270000,
}
```

```python
trials = df[df['is_trial'] == 1].groupby(df['month'])['user_id'].nunique()
trials
```

```python
cpa = {month: round(marketing_spend[str(month)] / trials.get(month), 2) for month in trials.index}
cpa
```

##CR

```python
trials_set = df[df['is_trial'] == 1][['user_id', 'month']].drop_duplicates()
first_pay = df[(df['is_trial'] == 0) & (df['renewal_number'] == 1)][['user_id']]
```

```python
merged = trials_set.merge(first_pay, on='user_id', how='left', indicator=True)
merged
```

```python
merged['converted'] = merged['_merge'] == 'both'
cr = merged.groupby('month')['converted'].mean()
```

##Rebill

```python
steps = df.groupby(['user_id'])['renewal_number'].max().reset_index()
steps
```

```python
df.groupby('user_id')['renewal_number'].max().max()
```

```python
#обчислення step-конверсій по місяцях
def calculate_step_transitions_by_month(df, max_step=47):
    df_clean = df[df['is_refunded'] == 0].copy()

    result = {}

    for month, group in df_clean.groupby('month'):
        user_max = group.groupby('user_id')['renewal_number'].max()
        month_result = {}

        for step in range(1, max_step + 1):
            with_prev = (user_max >= step - 1).sum()
            with_next = (user_max >= step).sum()
            month_result[f'step{step-1}_{step}'] = (with_next / with_prev) if with_prev else None

        result[month] = month_result

    return result
```

```python
# розрахунок ребіл на основі кроків
def calculate_rebill(step_rates, target_weeks=47):
    known_steps = list(step_rates.values())
    known_steps = [x for x in known_steps if x is not None]

    if not known_steps:
        return 1.0  # тільки trial

    if len(known_steps) < (target_weeks - 1):
        forecast_value = np.mean(known_steps[-3:])
        known_steps += [forecast_value] * ((target_weeks - 1) - len(known_steps))

    rebill = 1
    product = 1
    for step in known_steps:
        product *= step
        rebill += product
    return rebill
```

```python
step_transitions_by_month = calculate_step_transitions_by_month(df, max_step=47)

rebill_by_month = {
    str(month): round(calculate_rebill(steps), 2)
    for month, steps in step_transitions_by_month.items()
}

# Вивід
for month, rebill in rebill_by_month.items():
    print(f"{month}: Rebill = {rebill}")
```

##Price

```python
def get_price(pn):
    if '6.99' in pn:
        return 6.99
    elif '7.99' in pn:
        return 7.99
    else:
        return 0
```

```python
df['price'] = df['purchase_name'].apply(get_price)
```

```python
price = df[df['is_trial'] == 1].groupby('month')['price'].mean().round(2)
price
```

##LTV

```python
ltv = {}
for month in trials.index:
    ltv[str(month)] = round(price.get(month, 0) * cr.get(month, 0) * rebill_by_month.get(str(month), 0), 2)
```

```python
ltv
```

```python
result = pd.DataFrame({
    'CPA': pd.Series(cpa),
    'CR': pd.Series(cr),
    'Price': pd.Series(price),
    'Rebill': pd.Series(rebill_by_month),
    'LTV': pd.Series(ltv)
})

result.index = result.index.astype(str)
result = result.groupby(result.index).first()
result
```
